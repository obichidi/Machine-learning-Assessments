       'County', 'Zip', 'Lat', 'Lng', 'Population','TimeZone', 'Job':  I kept area to represent the demographics of the patient.

```python
df = df[['Area', 
       'Children', 'Age', 'Income', 'Marital', 'Gender', 'ReAdmis',
       'VitD_levels', 'Doc_visits', 'Full_meals_eaten', 'vitD_supp',
       'Soft_drink', 'Initial_admin', 'HighBlood', 'Stroke',
       'Complication_risk', 'Overweight', 'Arthritis', 'Diabetes',
       'Hyperlipidemia', 'BackPain', 'Anxiety', 'Allergic_rhinitis',
       'Reflux_esophagitis', 'Asthma', 'Services', 'Initial_days',
       'TotalCharge', 'Additional_charges', 'TimelyAdmission',
       'Timely_treatment', 'Timely_visits', 'Reliability', 'Options',
       'Hours_of_treatment', 'Couteous staff', 'Evidence of active listening']]
```

describe the data:

create dummys of non catagorical columns

```python
def get_dummys(df):
 
  for col in df:
    if not pd.api.types.is_numeric_dtype(df[col]):
      df = df.join(pd.get_dummies(df[col], prefix = col, drop_first= True))


  df = df.select_dtypes(np.number)
  return df

df = get_dummys(df)
df
```

```python
df.to_csv('/content/drive/MyDrive/d208_assessment/randon_forest_prepared.csv')
```

createa X variablet that holds the dataframe minus the dependant variable create a variable y and hold the bindependant variable.

```python
X = df.drop(columns=['ReAdmis_Yes'],axis=1)
y = df['ReAdmis_Yes']
```

preform a train test split to split the data into training an testing sets.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0 ,stratify=y)
```

```python
X_train.to_csv('/content/drive/MyDrive/d208_assessment/X_train.csv')
X_test.to_csv('/content/drive/MyDrive/d208_assessment/X_test.csv')
y_train.to_csv('/content/drive/MyDrive/d208_assessment/y_train.csv')
y_test.to_csv('/content/drive/MyDrive/d208_assessment/y_test.csv')
```

create a RandomForestClassifier model fit it and check the score:

```python
model = RandomForestRegressor(n_estimators=50)
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

store predictions in a variable:

```python
y_preds = model.predict(X_test)

```

check cmean squared error, mean absolute error:

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_preds))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_preds))
```
